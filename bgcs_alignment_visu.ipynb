```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mmseqs_db/result.tsv', sep='\t', header=None, names=['query', 'target', 'evalue', 'pident', 'qcov','bits' ])
```

```python
df = df[df['query'] != df['target']]  # Remove self-alignments]
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["pident"], bins=50, edgecolor="black")
plt.xlabel("% Identity")
plt.ylabel("Count")
plt.title("Identity distribution (all-vs-all)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["qcov"], df["pident"], alpha=0.3, s=5)
plt.xlabel("Query coverage (%)")
plt.ylabel("% Identity")
plt.title("Identity vs Coverage")
plt.tight_layout()
plt.show()
```
